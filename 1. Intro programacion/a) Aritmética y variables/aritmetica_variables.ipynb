## Variables
Hasta ahora, has utilizado código para realizar un cálculo e imprimir el resultado, y el resultado no se guarda en ningún lugar. Sin embargo, puedes imaginar que quizás quieras guardar el resultado para trabajar con él más tarde. Para ello, necesitarás utilizar variables.


### Creación de variables
La siguiente celda de código crea una variable llamada ``test_var`` y le asigna el valor que obtenemos al sumar 5 a 4.


```python
test_var = 5+4
```

```python
print(test_var)
```

```python
test_var
```

Luego imprimimos el valor que se le asigna a la variable, que es 9.

En general, para trabajar con una variable, debes comenzar seleccionando el nombre que quieres usar. Los nombres de las variables son idealmente cortos y descriptivos. También deben cumplir varios requisitos:

- No pueden tener espacios (p. ej., `test var` no está permitido)

- Solo pueden incluir letras, números y guiones bajos (p. ej., `test_var!` no está permitido)

- Deben comenzar con una letra o un guión bajo (p. ej., `1_var` no está permitido)

- Luego, para crear la variable, debes usar `=` para asignar el valor que quieres que tenga.

Siempre puedes echar un vistazo al valor asignado a la variable usando print() y poniendo el nombre de la variable entre paréntesis.
Con el tiempo, aprenderás a seleccionar buenos nombres para las variables de Python. ¡Está bien que te sientas incómodo ahora, y la mejor manera de aprender es simplemente viendo mucho código de Python!

### Uso de múltiples variables

Es habitual que el código utilice múltiples variables. Esto resulta especialmente útil cuando tenemos que realizar un cálculo largo con múltiples entradas.

En la siguiente celda del código, calculamos la cantidad de segundos en cuatro años. Este cálculo utiliza cinco entradas.


```python
año = 365
dia = 24
hora = 60 
minuto = 60
seg_en_4años = (año*dia*hora*minuto)*4
seg_en_4años
```

Como se calculó anteriormente, hay 126144000 segundos en cuatro años.

Tenga en cuenta que es posible realizar este cálculo sin variables como 60 * 60 * 24 * 365 * 4, pero es mucho más difícil comprobar que el cálculo sin variables no tiene algún error, porque no es tan legible. Cuando utilizamos variables (como num_years, days_per_year, etc.), podemos realizar un mejor seguimiento de cada parte del cálculo y comprobar y corregir más fácilmente los errores.

Tenga en cuenta que es particularmente útil utilizar variables cuando los valores de las entradas pueden cambiar. Por ejemplo, digamos que queremos mejorar ligeramente nuestra estimación actualizando el valor de la cantidad de días en un año de 365 a 365,25, para tener en cuenta los años bisiestos. Luego podemos cambiar el valor asignado a days_per_year sin cambiar ninguna de las otras variables y volver a hacer el cálculo.


```python
año = 365.25
dia = 24
hora = 60 
minuto = 60
seg_en_4años = (año*dia*hora*minuto)*4
seg_en_4años
```

Nota: Es posible que hayas notado que se agregó .0 al final del número, lo que puede parecer innecesario. Esto se debe a que en el segundo cálculo, usamos un número con una parte fraccionaria (365,25), mientras que en el primer cálculo se multiplicaron solo números sin parte fraccionaria. Aprenderás más sobre esto en la Lección 3, cuando tratemos los tipos de datos.
