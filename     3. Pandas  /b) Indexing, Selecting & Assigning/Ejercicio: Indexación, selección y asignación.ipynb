Pista: puedes usar `loc` o `iloc`. Cuando trabajes en la respuesta a esta pregunta y a las varias que le siguen, ten en cuenta el siguiente "escollo" descrito en el tutorial:

- `iloc` usa el esquema de indexación de la biblioteca estándar de Python, donde el primer elemento del rango está incluido y el último está excluido. `loc`, por su parte, indexa de forma inclusiva.

- Esto es particularmente confuso cuando el índice del DataFrame es una lista numérica simple, por ejemplo, 0,...,1000. En este caso, `df.iloc[0:1000]` devolverá 1000 entradas, mientras que `df.loc[0:1000]` devolverá 1001. Para obtener 1000 elementos usando loc, tendrás que pedir un elemento menos y solicitar `df.loc[0:999]`.
